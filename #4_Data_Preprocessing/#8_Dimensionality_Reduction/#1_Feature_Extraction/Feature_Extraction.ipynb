# Dimensionality Reduction: Feature Extraction


Feature Extraction Techniques:
- Principal Component Analysis ( PCA )
- Linear Discriminant Analysis (LDA)

## Upload Needed Libraries


```python
# ---------------- GENERAL -------------
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.preprocessing import StandardScaler
# ---------------- USED WITH PCA -----------
from sklearn.decomposition import PCA
# ----------------- USED WITH LDA ------------
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

## Principal Component Analysis ( PCA )

```python
# Step 1: Generate a sample dataset with 5 features
X, y = make_classification(n_samples=100, n_features=5, random_state=42)
```

```python
# Step 2: Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Step 3: Apply PCA to reduce to 2 components
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
```

```python
# Step 4: Visualize the results
plt.figure(figsize=(16, 6))

# Plot before PCA (original data - first 2 features)
plt.subplot(1, 2, 1)  # 1 row, 2 columns, 1st plot
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis', edgecolor='k', s=100)
plt.title('Original Data: First 2 Features')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')

# Plot after PCA
plt.subplot(1, 2, 2)  # 1 row, 2 columns, 2nd plot
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis', edgecolor='k', s=100)
plt.title('PCA Result: Reduced to 2 Components')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')

plt.tight_layout()
plt.show()
```

## Linear Discriminant Analysis (LDA)

```python
# Step 1: Generate a dataset with 5 features and 3 classes
X, y = make_classification(n_samples=100, n_features=5, n_classes=3, n_informative=3, random_state=42)
```

```python
# Step 2: Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Step 3: Apply LDA to reduce to 2 dimensions
lda = LinearDiscriminantAnalysis(n_components=2)
X_lda = lda.fit_transform(X_scaled, y)
```

```python
# Step 4: Visualize the results
plt.figure(figsize=(16, 6))

# Plot before LDA (original data - first 2 features)
plt.subplot(1, 2, 1)  # 1 row, 2 columns, 1st plot
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='rainbow', edgecolor='k', s=100)
plt.title('Original Data: First 2 Features')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')

# Plot after LDA (transformed data)
plt.subplot(1, 2, 2)  # 1 row, 2 columns, 2nd plot
plt.scatter(X_lda[:, 0], X_lda[:, 1], c=y, cmap='rainbow', edgecolor='k', s=100)
plt.title('LDA Result: Reduced to 2 Components')
plt.xlabel('Linear Discriminant 1')
plt.ylabel('Linear Discriminant 2')

plt.tight_layout()
plt.show()
```
