# Plotting & Data Visualization

## Including libraries

```python
# Plotting & visualization libraries
import matplotlib.pyplot as plt      # Matplotlib library abbreviated as 'plt'
import seaborn as sns                # Seaborn library abbreviated as 'sns'

# other helping libraries 
import numpy as np
import pandas as pd
```

# Matplotlib

### General 

#### **plt.show()** 
used to display all the open figures 

Takes no parameters

#### **plt.title()**
used to add title to the plot

Parameters : similar to xlabel & ylabel 

#### **plt.xlabel()** & **plt.ylabel()**
used to set the labels for the x and y axes respectively

Parameters:

Main:

- **xlabel:** (Type: str) The text to display as the x-axis label.

Optional:

- **fontdict:** (Type: dict) A dictionary controlling the appearance of the label text, such as font size, weight, family, etc. (Example: {'fontsize': 14, 'fontweight': 'bold', 'color': 'red'})
- **labelpad:** (Type: float) The spacing between the label and the x-axis, in points. 
- **loc:** (Type:'left', 'center', 'right') The alignment of the label relative to the x-axis.


#### **plt.grid ()**  

used to add or modify grid lines on a plot

plt.grid(b=**False**, which=**'major'**, axis=**'both'**, color=**'k'**, linestyle=**'-'**, linewidth=0.5, alpha=**None**)


Parameters (optional):
- **b:** (Type: bool) Controls whether grid lines are shown or hidden, True turns on the grid.
- **which:** (Type: 'major', 'minor', 'both') Specifies which grid lines to show.

      > 'major': Shows grid lines for major ticks.

      > 'minor': Shows grid lines for minor ticks.

      > 'both': Shows grid lines for both major and minor ticks.

- **axis:** (Type: 'both', 'x', 'y') Specifies which axis to apply the grid lines to.

      > 'both': Applies grid lines to both x and y axes.

      > 'x': Applies grid lines to the x-axis only.

      > 'y': Applies grid lines to the y-axis only.

- **color:** Sets the color of the grid lines. Accepts any Matplotlib color format (e.g., 'r', '#00FF00', etc.)
- **linestyle:** (Type: str) Sets the style of the grid lines

    Common styles include:

      > '-': Solid line

      > '--': Dashed line

      > '-.': Dash-dot line

      > ':': Dotted line

- **linewidth:** (Type: float) Sets the width of the grid lines.
- **alpha:** (Type: float) Sets the transparency of the grid lines


### Histogram

#### **plt.hist ( )**
plt.hist( x, bins=**None** , range=**None**, density=**False**, weights=**None**, cumulative=**False**, bottom=**None**, histtype=**'bar'**, align=**'mid'**, orientation=**'vertical'**)

**Parameters:**

Main:

- **x:** data to be plotted as a histogram

Optional:

- **bins:** Specifies the number of equal-width bins in the range.
- **range:** (Type: tuple or none) takes a tuple specifying the lower and upper range of the bins. Values outside this range are ignored.
- **density:** (Type: bool) If True, the result is normalized to form a probability density, meaning the area under the histogram will sum to 1.
- **weights:** An array of weights, of the same shape as x. Each value in x only contributes its associated weight towards the bin count.
- **cumulative:** (Type: bool) If True, the histogram is cumulative, where each bin gives the counts up to that point in the data.
- **bottom:** (Type: array-like, scalar, or None) A location on the y-axis to start drawing bars from, useful for stacked histograms.
- **histtype:** The type of histogram to draw.(Type: 'bar', 'barstacked', 'step', 'stepfilled')

      > 'bar': Traditional bar-type histogram.

      > 'barstacked': Bar-type histogram with multiple data stacked on top of each other.  

      > 'step': Generates a lineplot that is unfilled.

      > 'stepfilled': Generates a filled lineplot.
      
- **align:** The alignment of the bars. 'mid' aligns the bars centrally with the bin edges  (Type: 'left', 'mid', 'right' )
- **orientation:** ( Type: 'vertical', 'horizontal' ) Vertical bars have the height represent the data, while horizontal bars have the width represent the data.



```python

# Random generated data
data = np.random.randn(1000)

# Create a histogram
plt.hist(
    data,
    bins=30,
    range=(-3, 3),
    density=True,
    color='red',
    alpha=0.7,
    histtype='bar',
    edgecolor='black',
    rwidth=0.9
)

# Adding labels and title
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Histogram of Normally Distributed Data')

# Show the plot
plt.show()
```

### Scatter plot

#### **plt.scatter ( )**
plt.scatter(x, y, s=**None**, c=**None**, marker=**None**, cmap=**None**, norm=**None**, vmin=**None**, vmax=**None**, alpha=**None**, linewidths=**None**, edgecolors=**None**)


**Parameters:**

Main:

- **x:** x-coordinates of the data points.
- **y:** y-coordinates of the data points.

Optional:

- **s:** (Type: scalar or array-like ) The size of the markers. A scalar applies the same size to all markers, while an array applies different sizes to each point.
- **c:** The color of the markers. This can be a single color or an array of colors for each point. 
- **marker:** (Type: str) The shape of the markers. Common options include 'o' for circles, '^' for triangles, and 's' for squares.
- **cmap:** A colormap to map numeric data to colors when c is an array.
- **norm:** Normalization used to scale luminance data to [0, 1].
- **vmin and vmax:** (scalar) Values for normalization of colormap. If not set, the minimum and maximum of the colormap are used.
- **alpha:** (Type: float) The transparency level of the markers. 
- **linewidths:** (Type: scalar or array-like) The width of the marker edges. Can be a single width for all points or an array of widths.
- **edgecolors:**  The color of the marker edges. Can be set to 'none', 'face', or specific color(s).

```python

# Generating random data 
x = np.random.rand(50)
y = np.random.rand(50)

# Create a scatter plot
plt.scatter(
    x, 
    y, 
    s=50, 
    c='blue', 
    alpha=0.5, 
    edgecolors='black', 
    linewidths=1.5
)

# Adding labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot Example')


# Show the plot
plt.show()
```

### Box plot

#### **plt.boxplot( )**

plt.boxplot(x, notch=**False**, vert=**True**, patch_artist=**False**, widths=**None**, meanline=**False**, showmeans=**False**, showcaps=**True**, showbox=**True**, whis=**1.5**, showfliers=**True**, flierprops=**None**, boxprops=**None**, capprops=**None**, whiskerprops=**None**)

**parameters:**

Main:
- **x:** The input data, which can be a single dataset or multiple datasets. Each dataset will generate a separate box plot.

Optional
- **notch:** (Type: bool) If True, the box plot will have a notch at the median. Notches provide a rough guide to significance of difference between medians.
- **vert:** (Type: bool) If True, plots vertical boxes. If False, plots horizontal boxes.
- **patch_artist:** (Type: bool) If True, fills the box with color.
- **widths:** (Type: scalar or array-like) Width of the boxes. A scalar applies the same width to all boxes, while an array specifies widths individually.
- **meanline:** (Type: bool) If True, will draw a line for the mean instead of a point.
- **showmeans:** (Type: bool) If True, will show the mean of the data.
- **showcaps:** (Type: bool) If True, will show the caps at the ends of the whiskers.
- **showbox:** (Type: bool) If True, will display the box.
- **whis:** (Type: float, sequence, or str)

      > If a float, the whis parameter determines the reach of the whiskers to be a multiple of the IQR.

      > If a sequence of two numbers, the whiskers extend to the indicated percentiles.

      > If the string 'range', the whiskers extend to the min and max of the data.

- **showfliers:** (Type: bool) If True, outlier points are shown.
- **flierprops:** (Type: dict) A dictionary of style specifications for the outliers. Can specify marker style, color, and size.
- **boxprops:** (Type: dict) A dictionary of style specifications for the box. Can specify color, linewidth, etc.
- **capprops:** (Type: dict) A dictionary of style specifications for the caps.
- **whiskerprops:** (Type: dict) A dictionary of style specifications for the whiskers.




```python

# Sample data: Three different groups
data = np.random.rand(100)

# Create a box plot
plt.boxplot(
    data,
    vert=True,   # Vertical boxes
    patch_artist=True,  # Fill boxes with color
    widths=0.4,  # Width of boxes
    meanline=True,  # Show mean as a line
    showmeans=True,  # Display means
    boxprops=dict(facecolor='lightblue', color='blue'),  # Style for box
    whiskerprops=dict(color='green', linestyle='--'),  # Style for whiskers
    flierprops=dict(marker='o', color='purple', markersize=5),  # Style for outliers
    capprops=dict(color='blue')  # Style for caps
)

# Adding labels and title
plt.ylabel('Values')
plt.title('Box Plot Example')

# Show the plot
plt.show()
```

### Bar plot

#### **plt.bar()**
plt.bar (x, height=**None**, width=**None**,bottom=**None**, align=**'center'**, colour, edgecolour, linewidth, tick_label  )

**Parameters**

Main:

- **x**: The data points to be plotted, specifying the categories.

Optional:

- **height**: The height of the bars. 
- **width**: (Type: scalar or array-like) The width of the bars. It can be a single value applied to all bars or an array specifying the width of each bar individually.
- **bottom:** (Type: scalar or array-like) The y-coordinate(s) of the bottom of the bars. This can be used to create stacked bar plots.
- **align:** (Type: 'center', 'edge') Specifies the alignment of the bars with respect to the x-ticks. 
- **color:** (Type: color or list of colors) The color of the bars. It can be a single color for all bars or a list of colors for individual bars.
- **edgecolor:** (Type: color or list of colors) The color of the bar edges.
- **linewidth** (Type: scalar or array-like) The width of the bar edges.
- **tick_label:** (Type: string or array-like) The labels for the bars. If not provided, the labels are taken from x.




```python

# Sample data
categories = ['A', 'B', 'C', 'D']
values = [4, 7, 1, 8]

# Create a bar plot
plt.bar(
    categories, 
    values, 
    width=0.6, 
    color=['blue', 'green', 'red', 'purple'], 
    edgecolor='black', 
    linewidth=1.5, 
    tick_label=['Group A', 'Group B', 'Group C', 'Group D'], 
    align='center', 
    capsize=5
)

# Add labels and title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot Example')

# Display the plot
plt.show()
```

# Seaborn

Seaborn is built on top of Matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics. It is used for creating complex plots with minimal code and effort.

### Histogram

#### **sns.histplot( )**

sns.histplot(data, x=**None**, y=**None**, hue=**None**, weights=**None**, stat=**'count'**, bins=**'auto'**, binwidth=**None**, binrange=**None**, cumulative=**False**, multiple=**'layer'**, element=**'bars'**, shrink=**None**, palette=**None** , colour=**None**, alpha=**None**)

**Parameters**

Main:

- **data**: (Type: DataFrame, array, or list of arrays) The dataset from which variables will be extracted for plotting.

Optional:

- **x** (Type: str or vector): The data or variable name to be plotted on the x-axis.
- **y** (Type: str or vector): If provided, creates a bivariate histogram (histogram in two dimensions).
- **hue** (Type: str or vector): Variable that defines subsets of data, which are plotted with different colors.
- **weights:** (Type: vector) If provided, the histogram bins are weighted by this array of values.
- **stat:** (Type:‘count’, ‘frequency’, ‘density’, ‘probability’) Determines the statistic that is computed and represented on the y-axis

    > 'count': Show the count of observations in each bin.

    > 'frequency': Show the frequency of observations.

    > 'density': Normalize the bars so that the area sums to 1.

    > 'probability': Normalize the bars so that they sum to 1.

- **bins:** (Type: int or sequence) Number of bins or a sequence of bin edges to use.
- **binwidth:** (Type: float) Width of each bin.
- **binrange:** (Type: tuple): The minimum and maximum values for the bins.
- **cumulative:** (Type: bool) If True, displays the cumulative histogram.
- **multiple:** (Type:‘layer’, ‘dodge’, ‘stack’, ‘fill’) Determines how multiple distributions are drawn

    > 'layer': Overlapping distributions.

    > 'dodge': Separate distributions side by side.

    > 'stack': Stacked distributions.

    > 'fill': Normalized so that the stack height is 1.

- **shrink:** (Type: float) Scale factor to adjust the width of each bar (useful for dodging).
- **element:** (Type: ‘bars’, ‘step’, ‘poly’) Type of visual representation for each bin

    > 'bars': Traditional histogram bars.

    > 'step': Step plot.

    > 'poly': Polygonal fill between steps.

- **palette:** (Type: string, list, dict, or Colormap) Colors to use for different levels of the hue variable.
- **color:** (Type: matplotlib color) Color for all elements.
- **alpha:** (Type: float) Transparency level of the histogram bars.



```python

# Generate random data
data = np.random.randn(1000)

# Create a histogram
sns.histplot(data, bins=30, color='green', stat='density', alpha=0.6)

# Add labels and title
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Histogram with Density Plot')

# Show the plot
plt.show()
```

### Scatter plot

#### **sns.scatterplot()**

sns.scatterplot(x=**None**, y=**None**, data =**None**, hue=**None**, style=**None**, size=**None**, palette=**None**, markers=**True**,  alpha=**None**)

**Parameters:**

Main:

- **x:** Values to plot on the x-axis.
- **y:** Values to plot on the y-axis.
- **data:** (Type: DataFrame) The data source containing the variables to plot.

Optional:

- **hue:** Variable in the data to map plot aspects to different colors.
- **style:** (Type: variable) Variable to map plot aspects to different marker styles.
- **size:** (Type: scalar or array ) Variable to map plot aspects to different sizes of markers.
- **palette:** (Type: string, list, dict, or Colormap) Method for choosing the colors to represent different hue levels.
- **marker:** (Type: str) The shape of the markers. Common options include 'o' for circles, '^' for triangles, and 's' for squares.
- **alpha:** (Type: float) The transparency level of the markers. 



```python

# Generate random data
n = 100
data = pd.DataFrame({
    'x': np.random.rand(n) * 100,  # Random x-coordinates
    'y': np.random.rand(n) * 100,  # Random y-coordinates
    'category': np.random.choice(['A', 'B'], n),  # Random categories
    'size': np.random.rand(n) * 1000  # Random sizes for markers
})

# Create a scatter plot
sns.scatterplot(
    data=data,            # The DataFrame to use
    x='x',                # x-coordinates
    y='y',                # y-coordinates
    hue='category',       # Color by category
    size='size',          # Size of markers
    palette='viridis',    # Color palette
    style='category',     # Marker style by category
    markers=['o', 's'],   # Marker shapes for different styles
    alpha=0.6             # Transparency level of markers
)

# Show plot
plt.title('Scatter Plot with Random Data')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.show()
```

### Box plot

#### **sns.boxplot()**

**Parameters:**

Main:

- **data:** (Type: DataFrame) The dataset to plot, which can be a DataFrame, array, or list of arrays.
- **x:** (Type: string or array-like) The name of the column or the array for x-coordinates.
- **y:** (Type: string or array-like) The name of the column or the array for y-coordinates.

Optional:

- **hue:** (Type: string or array-like) Variable in the data to map plot color to. 
- **order:** (Type: list of strings) Order to plot the categorical levels along the x-axis.
- **hue_order:** (Type: list of strings) Order to plot the categorical levels of the hue variable.
- **orient:** (Type: string) Orientation of the plot. Can be 'v' (vertical) or 'h' (horizontal).
- **color:** (Type: matplotlib color) Color for all elements, or use palette for multiple colors.
- **palette:** (Type: string, list, or dict) Colors to use for different levels of the hue variable.
- **saturation:** (Type: float) Proportion of the original saturation to draw colors at.
- **width:** (Type: float) Width of each box as a proportion of the available space.
- **dodge:** (Type: bool) Whether to separate boxes by hue levels along the categorical axis.
- **fliersize:** (Type: float) Size of the markers used to represent outliers.
- **linewidth:** (Type: float) Width of the lines that frame the plot elements.


```python


# Generate random data
n = 100
data = pd.DataFrame({
    'Value': np.random.randn(n) * 10 + 50  # Random values with mean 50 and std deviation 10
})

# Create a box plot
sns.boxplot(
    data=data,         # The DataFrame to use
    y='Value',         # Numeric variable for the y-axis
    color='skyblue',   # Color of the box
    width=0.4,         # Width of the box
    fliersize=5        # Size of the outliers
)

#Add grid
plt.grid(True, linestyle='--', alpha=0.7)

# Show plot
plt.title('Single Box Plot with Random Data')
plt.ylabel('Value')
plt.show()
```

### Bar plot

#### **sns.barplot ( )**

**Parameters:**

Main:

- **data:** (Type: DataFrame) The dataset to plot, which can be a DataFrame, array, or list of arrays.
- **x:** (Type: string or array-like) The name of the column or the array for x-coordinates (categories).
- **y:** (Type: string or array-like) The name of the column or the array for y-coordinates (values).

Optional:

- **hue:** (Type: string or array-like) Variable in the data to map plot color to. It can be a column name or array-like values.
- **order:** (Type: list of strings) Order to plot the categorical levels along the x-axis.
- **hue_order:** (Type: list of strings) Order to plot the categorical levels of the hue variable.
- **orient:** (Type: string) Orientation of the plot. Can be 'v' (vertical) or 'h' (horizontal).
- **color:** (Type: matplotlib color) Color for all elements, or use palette for multiple colors.
- **palette:** (Type: string, list, or dict) Colors to use for different levels of the hue variable.
- **saturation:** (Type: float) Proportion of the original saturation to draw colors at.
- **dodge:** (Type: bool) Whether to separate bars by hue levels along the categorical axis.

```python

# Generate random data
n_categories = 5
data = pd.DataFrame({
    'Category': [f'Category {i}' for i in range(1, n_categories + 1)],
    'Value': np.random.randint(10, 100, n_categories),                       # Random integer values between 10 and 100
    'Group': np.random.choice(['Group 1', 'Group 2'], n_categories)          # Randomly assign groups
})

# Create a customized bar plot
sns.barplot(
    data=data,
    x='Category',          # Categorical variable for the x-axis
    y='Value',             # Numeric variable for the y-axis
    hue='Group',           # Grouping variable for color
    palette='Paired',      # Color palette
    saturation=0.8,        # Saturation of colors
    dodge=True,            # Separate bars by hue levels
    edgecolor='black',     # Color of the bar edges
    linewidth=1.5,         # Line width of bar edges
    
)

# Customize plot appearance
plt.title('Customized Bar Plot with Random Data', fontsize=14, fontweight='bold')
plt.xlabel('Category', fontsize=12)
plt.ylabel('Value', fontsize=12)

plt.show()
```

### HeatMap

#### **sns.heatmap ( )**

sns.heatmap(data, vmin=**None**, vmax=**None**, cmap=**None**, center=**None**, robust=**False**, annot=**None**, fmt=**'.2g'**, linewidths=**None**, linecolor=**'white'**, cbar=**True**)

**Parameters:**

Main:

- **data:** (Type: 2D array-like or DataFrame) Rectangular data to plot as a heatmap.

Optional:

- **vmin and vmax:** (Type: scalar) Values to anchor the colormap, otherwise they are inferred from the data and other keyword arguments.
- **cmap:** (Type: matplotlib colormap) The colormap to use for mapping values to colors.
- **center:** (Type: float) Value at which to center the colormap when plotting divergent data.
- **robust:** (Type: bool) If True, ignores data points above and below the 95th and 5th percentiles when calculating vmin and vmax.
- **annot:** (Type: bool or rectangular dataset) If True, write the data value in each cell.
- **fmt:** (Type: string) String formatting code to use when adding annotations.
- **linewidths:** (Type: float) Width of the lines that will divide each cell.
- **linecolor:** (Type: color) Color of the lines that will divide each cell.
- **cbar:** (Type: bool) Whether to draw a colorbar.


```python

# Generate random data
data = np.random.rand(10, 12)  # Random data for a 10x12 matrix

# Create a Pandas DataFrame with row and column labels
df = pd.DataFrame(
    data,
    index=[f'Row {i}' for i in range(1, 11)],
    columns=[f'Col {i}' for i in range(1, 13)]
)

# Create a customized heatmap
sns.heatmap(
    data=df,
    vmin=0, vmax=1,          # Set limits for the colormap
    cmap='coolwarm',         # Colormap to use
    center=0.5,              # Center the colormap at 0.5
    annot=True,              # Annotate each cell with its value
    fmt='.2f',               # Format annotation with two decimal places
    linewidths=0.5,          # Width of the lines dividing cells
    linecolor='black',       # Color of the lines dividing cells
    cbar=True,               # Include a colorbar
    
)

# Customize plot appearance
plt.title('Customized Heatmap Example', fontsize=14, fontweight='bold')
plt.xlabel('Columns', fontsize=12)
plt.ylabel('Rows', fontsize=12)
plt.show()
```
